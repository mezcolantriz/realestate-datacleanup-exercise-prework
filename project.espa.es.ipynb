```python
# Definir las columnas de interés
columnas_interes = ['price', 'rooms', 'surface', 'bathrooms']

#quiero que quede más entendible en español así que cambio los nombres de las variables
nombres_personalizados = {
    'price': 'Precio',
    'rooms': 'Habitaciones',
    'surface': 'Superficie',
    'bathrooms': 'Baños'
}

# Rellenar NaN con 0 en las columnas de interés
cinturon_sur_ds_filled = cinturon_sur_ds[columnas_interes].fillna(0)

# Calcular la media y varianza muestral para todo el Cinturón Sur
print("Cálculos para el total del Cinturón Sur: \n")

# Iterar sobre las columnas de interés
for columna in columnas_interes:
    # Calcular la media y redondear a 2 decimales
    media_columna = cinturon_sur_ds_filled[columna].mean().round(2)
    
    # Calcular la varianza muestral y redondeo a 2 decimales también
    varianza_columna = cinturon_sur_ds_filled[columna].var().round(2)

   # Imprimir los resultados usando los nombres personalizados
    print(nombres_personalizados[columna])  
    print(f"Media: {media_columna}")
    print(f"Varianza muestral: {varianza_columna}")
    print("-" * 30, "\n")
```

- Podemos ver como en Alcorcón, Fuenlabra y Getafe se da una alta concentración en la normalización más baja de precios, 
  mientras Leganés aún teniendo una alta concentración de baja normalización tiene altas normalizaciones más frecuentes que en el resto de poblaciones.

```python
# Eliminar filas con valores nulos en 'price' o 'surface'
cinturon_sur_ds_clean = cinturon_sur_ds.dropna(subset=['price', 'surface'])

# Calcular 'pps' en el DataFrame limpio
cinturon_sur_ds_clean.loc[:, 'pps'] = cinturon_sur_ds_clean['price'] / cinturon_sur_ds_clean['surface']

# Filtrar los municipios de interés Getafe y Alcorcón
getafe = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == 'Getafe']
alcorcon = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == 'Alcorcón']

# Calcular estadísticas para Getafe
getafe_pps_mean = getafe['pps'].mean()
getafe_pps_median = getafe['pps'].median()
getafe_pps_std = getafe['pps'].std()

# Calcular estadísticas para Alcorcón
alcorcon_pps_mean = alcorcon['pps'].mean()
alcorcon_pps_median = alcorcon['pps'].median()
alcorcon_pps_std = alcorcon['pps'].std()

# Imprimir los resultados
print(f"\nGetafe - Precio por m²: Media = {getafe_pps_mean:.2f}, Mediana = {getafe_pps_median:.2f}, Desviación estándar = {getafe_pps_std:.2f}")
print(f"Alcorcón - Precio por m²: Media = {alcorcon_pps_mean:.2f}, Mediana = {alcorcon_pps_median:.2f}, Desviación estándar = {alcorcon_pps_std:.2f}")
```

- Alcorcón tiene precios por metro cuadrado algo más altos en general que Getafe, tanto en la media como en la mediana. Las casas serán más caras según m2.

```python
#NO LO HE CONSEGUIDO :( NO PUC MES POR HOY :( )

from ipyleaflet import Map, Marker, WidgetControl
from ipywidgets import VBox, Label, Layout

# Crear el mapa centrado en Madrid
mapa = Map(center=(40.4, -3.7), zoom=12, layout=Layout(width="100%", height="600px"))

# Filtrar las 4 poblaciones (Fuenlabrada, Alcorcón, Getafe, Leganés)
fuenlabrada = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == "Fuenlabrada"]
alcorcon = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == "Alcorcón"]
getafe = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == "Getafe"]
leganes = cinturon_sur_ds_clean[cinturon_sur_ds_clean['level5'] == "Leganés"]

# Función para agregar marcadores sin personalizar
def agregar_marcadores(df, mapa):
    for index, row in df.iterrows():
        # Verificar las coordenadas de cada marcador
        #print(f"Marcador en: {row['latitude']}, {row['longitude']}")
        
        # Crear el marcador (sin icono personalizado)
        marker = Marker(location=(row['latitude'], row['longitude']))
        mapa.add_layer(marker)

# Añadir marcadores al mapa para cada población
agregar_marcadores(fuenlabrada, mapa)
agregar_marcadores(alcorcon, mapa)
agregar_marcadores(getafe, mapa)
agregar_marcadores(leganes, mapa)

# Crear una leyenda con VBox
leyenda = VBox([
    Label(value="Leyenda:"),
    Label(value="📍 Fuenlabrada (Azul)"),
    Label(value="📍 Alcorcón (Naranja)"),
    Label(value="📍 Getafe (Verde)"),
    Label(value="📍 Leganés (Rojo)")
], layout=Layout(width='200px', height='150px'))

# Añadir la leyenda al mapa usando WidgetControl
leyenda_control = WidgetControl(widget=leyenda, position='topright')
mapa.add_control(leyenda_control)

# Mostrar el mapa interactivo
mapa
```
